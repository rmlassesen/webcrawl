```python
from lib import *
import networkx as nx
import matplotlib.pyplot as plt
```

```python
spun = webcrawl.crawl('http://www.pythex.org', 4)
```

```python
spuntest = spun
```

```python
url_indexes = {}
url_index_count = 0
def nest_replace(dictionary, url_indexes, url_index_count):
    new_dict = {}
    i = -1
    for key, value in dictionary.items():
        if not key in url_indexes:
            url_index_count += 1
            url_indexes[key] = url_index_count
            new_dict[url_index_count] = None
            i += 1
    
    for key, value in dictionary.items():
        if isinstance(value, dict):
            new_dict[url_index_count - i], url_indexes, url_index_count = nest_replace(value, url_indexes, url_index_count)
    
    return new_dict, url_indexes, url_index_count
```

```python
spuntest, url_indexes, url_index_count = nest_replace(spuntest, url_indexes, url_index_count)
```

```python
from collections import Mapping
G = nx.DiGraph()

q = list(spuntest.items())
```

```python
spuntest
```

```python

while q:
    value, dictionary = q.pop()
    for new_value, new_dictionary in dictionary.items():  
        G.add_edge(value, new_value)
        if isinstance(new_dictionary, Mapping):
            q.append((new_value, new_dictionary))

```

```python
plt.figure(1,figsize=(16,16)) 
nx.draw(G, with_labels=True)
plt.savefig('networkx.png')
```
